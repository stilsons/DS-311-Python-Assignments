## 1. Green and Blue Fields

### Rows and Columns

Everything that you build in Tableau depends on whether we're using a green continuous field, or a blue discrete field. The first thing to understand is how we actually build a visualization, that is using drag and drop. So what we do is we drag and drop a continuous, or a discrete field onto our rows and columns and that will build the visualization. The combination of discrete continuous field determine the type of visualization that we get, whether it's a bar chart, a line chart, a pie chart, but to understand that we need to see the difference between using discrete and continuous field and Tableau has two very strict rules for this. The first rule, is for every continuous field that we put on a row or a column, we get an axis. Two important rule to remembers here,

1. Any green / continuous field add on the row or column will turns into an axis.

2. Any blue / discreet field add on the row or column will turns into header.

Example: 
- Drag sales out of the meausres and put it onto the column
- We get the sum of sales return in a single bar
- Drag profit onto column as well
- We have one for sum of sales and one for sum of profit
- Drag category field onto the rows
- We get a category header, which includes furniture, office, supplies, and technology
- We get three bars for each one of the two axes

![rows and columns](images/green_blue_1.png)

### Colors

Now lets introduce a little bit of color in Tableau. The way that Tableau handles color depends on whether we're coloring by a green or a blue field, that is a discrete or a continuous field.  So it's really important to only use color where it actually helps your analysis rather than hinder it. Tableau has a maximum with 20 distinct colors it can show in any one particular time. If you've got 21, it will cycle round and start reusing some of the colors. So that can be really bad if you're trying to use that as some kind of color key. So I'd always recommend keeping it around five or six. Any more than that, is color a viable thing to use?

Example:
- Drag profit field to column
- Drag category and subcategory fields to row
- Drag category field from data pane into color

![color](images/green_blue_2.png)

### Filters

The way that filtering works in Tableau is that you give it a measure or a dimension and then include or exclude certain values to filter those values out. And then Tableau will display the results of that filter. So the way I always think of it, is the left-hand side is the data pane. This is where all our data is. On the right-hand side is our view, the actual visualization of the data that we have. And then we have the little filters card between the two. So it's as if the data comes from over here on the left, is filtered through here in the middle, and that appears in our view.

Example:
- Drag profit field to column
- Drag category and subcategory fields to row
- Drag Category field from data pane into filter card
- Check off the category within the field to exclude them
- OR right click on the category field in the data pane and choose "Show Filter", then choose the cateogry to be included in the visualization

## 2. Comparing Measures

### Measure names and measure values

One of the real fundamental things to understand in Tableau is the difference between discrete and continuous fields. In Tableau, discrete fields are blue, continuous are green. In terms of the green and blue fields, the one we're interested in that moment in particular is what happens when we put them on the row and the column. Tableau has a very strict rule, and that is: 

Whenever you put a green field, such as sales, onto the row or column, you get an axis. And the rule is, for every field that is green, you get one axis. So if I added in a second field, I get two axes. 

But what about if we want to do something different? Like, combine these fields into a single axis. Can we do that? Well, not by these rules, because every green field gives us an axis. However, Tableau has got a little trick up its sleeve to let you break its own self-imposed rules. That is where measure names and measure values come from. 

Well, what are **measure names** and **measure values**, and where are they in the data? If you look at the data pane, we have our dimensions and we have our measures. Some are discrete and some continuous. But you'll notice, there's also some fields that are in italics, such as *measure names*, *latitude*, *longitude*, *orders*, and the *measure values*. They're in italics because these are fields that Tableau has generated itself, they're not part of the underlying data. Tableau generates these fields so that we can use them as and when we need to. 

One particular case is when we want to combine more than one measure into a view. Remember, when I drag sales and profit onto the column, it gave us two axes. Well, what happens if we drag measure values? We get a single bar, we get a single axis called value. But something else has happened. We've got a new card called measure values, and within it is each one of our measures. The measure values actually acts as a container, so that's what happens if we put measure values on the column. 

What about if we put measure names on the row, what's going to happen? Well, in the typical behavior that Tableau does, whenever we put a discrete field, it breaks all of the marks into their lower components. So in this case, putting the measure names on actually breaks that measure value bar into a separate bar for each one of our measures. This means that we can actually have more than one bar existing on the same axis. Have we broken any of Tableau's rules? No, because we still have a single green field on the column, and one single axis. But, what Tableau has done is by using the measures and measure values as a container, it's actually letting us visualize the container and then smash the container into its component parts. So not only is Tableau satisfied the rules, but it allowed us to draw the visualization that we want. 

If we want to remove some of these measure values, we can do it. In our visualization, discount is actually a percentage, and it makes no sense to sum it up. In addition, percentage and the value is much smaller than anything like the sales or the shipping costs. so it actually is insignificant. To remove it, we simply highlight it and drag it out of the container. Now, in this particular view, it doesn't really make sense because the units are all different. But the fundamental concept of being able to compare individual measures is very, very useful. Measure names and measure values will sometimes appear when you least expect them, but they're nothing to be afraid of. All they are is a container to allow us to bypass Tableau's own internal rules.

### Creating a multiple measure crosstab

One way of comparing measures is to actually use a crosstab. Now it doesn't always feel like a crosstab is actually data visualization. But when you think about it, it is. It's data and it's being displayed in visual form. Just because it doesn't look like a bar or a line chart doesn't mean it's not a good visualization. And in the right place and in the right time and for the right circumstances, a crosstab can actually be a really good choice, particularly if we're after the specific values and we want to see them easily. So how do we build a crosstab? Well, actually it's pretty straightforward when we understand how Tableau builds its charts. And when you look at the screen when we first open Tableau, connect to some data, you see that Tableau has got three drop field here labels, one for the column, one for the rows. And we've mostly been using those through this course, but there's a third place where we can drop something, right in the center of our view. Drop field here. 

Now, whenever we build a visualization, we've always been using this concept of we are aggregating a measure and then we're splitting it by dimension. And that's exactly the same thing with a crosstab, because in the case of a bar chart, the marks that we split up are bars. In a line chart, the mark types are lines. Well, in a crosstab, the mark types are text. So what we're in fact doing is we are splitting the text into smaller and smaller numbers. Let's have a look at that in action to fully understand it. 

So say we want to look at our sales. The first thing we do is we drag sales into our view. Now I'm not going to put it on the rows of the column. I'm going to just drop it right in the middle. And we get the mark of a mark type of text. And it works exactly the same way that the bars and the lines do, that if I add dimensions to the rows and the columns, it's going to subdivide this up. So if I put, say, subcategory on the rows, it's now divided all of the marks up by sub category. And the mark type is text, so it gives me the numeric value. If I then want it to subdivide this further, say by segment, I could put that on the column. And now I can see the individual values for the corporate chairs, for example. So I can read along, read down and see what I want. Now that's just looking at a single measure, it's looking at the sales. So how do we bring in another measure? Say if I wanted to see sales and profit at the same time? 

Well, if you think about when we did combined access charts, where we dragged and dropped the second field onto the access, it's kind of a similar principle. Now we didn't drag and drop sales onto our axis, we dropped it in the middle. So if we want to add in another measure, we have to drop it into the field again. So let's see what happens. So we can create crosstabs in exactly the same way as we create bar charts, because we subdivide down the values. But the advantage with the crosstab over other types of visualization is that you can actually see the numbers underneath. Is this the correct chart type to use? So if we want to look at the different scales and the magnitude, then maybe the combined bar or a bar in bar chart is a better use. But you need to know which chart to use in which situation. And in a certain situation, crosstabs can do a great job.

### Comparing two measures using a scatter plot

Now a scatterplot uses one of the key principles in Tableau. That is, for every continuous field you put on a column or row, you get an axis. And one of the key properties of a scatterplot is it has two axis, X and Y. So that tells us that to build one in Tableau, we need two measures, that are continuous, need one on the column and one on the right. So with my dataset, we'll have a look and see how sales and profit compare. So let's put our sales on one axis and we'll put our profit on the other. Okay, now this doesn't look like a scatterplot, but it is. Because you have to remember what Tableau will always do. Tableau will always try and aggregate the measures together. And what we're plotting here is actually the sum of profit and the sum of sales. And the total sum of profit and the total of sum of sales is one single mark. So that's why we've just got the single mark. But what we really want to see is how this mark can be subdivided up, using some of our dimensions. It's much in the same way that we have to break open a bar chart into its components. Same thing's true with a scatter plot. So what about if we want to compare, say, a subcategory.
